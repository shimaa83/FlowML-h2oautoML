```python
# Initialize H2O cluster
h2o.init()

# Load the CSV file into an H2O Frame
data = h2o.import_file('/kaggle/input/iris-flower-dataset/IRIS.csv')

# Split the data into training and testing sets
train, test = data.split_frame(ratios=[.8], seed=1234)

# Define the response and predictor variables
y = 'species'  # replace with your response column
x = data.columns
x.remove(y)

# Set up MLflow experiment
experiment_name = "H2O_AutoML_Experiment"
mlflow.set_experiment(experiment_name)

# Define different sets of parameters to experiment with
params_list = [
    {'max_runtime_secs': 1000, 'max_models': 5},
    {'max_runtime_secs': 1800, 'max_models': 10},
    {'max_runtime_secs': 2000, 'max_models': 15},
]

for params in params_list:
    with mlflow.start_run():
        # Log parameters
        mlflow.log_params(params)
        
        # Run H2O AutoML
        aml = H2OAutoML(max_runtime_secs=params['max_runtime_secs'], max_models=params['max_models'], seed=1234)
        aml.train(x=x, y=y, training_frame=train)

        # View the AutoML leaderboard
        lb = aml.leaderboard
        print(lb)

        # Log leaderboard as an artifact
        leaderboard_path = "leaderboard.csv"
        lb.as_data_frame().to_csv(leaderboard_path)
        mlflow.log_artifact(leaderboard_path)

        # Make predictions on the test set
        preds = aml.predict(test)

        # Evaluate the model on the test set
        test_performance = aml.leader.model_performance(test)
        mlflow.log_metric("mse", test_performance.mse())
        mlflow.log_metric("rmse", test_performance.rmse())
        mlflow.log_metric("r2", test_performance.r2())

        # Log the model
        mlflow.h2o.log_model(aml.leader, "model")

# Shut down the H2O cluster
h2o.shutdown()
```
